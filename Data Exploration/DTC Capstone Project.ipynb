```python
county_isotope_count = rd.groupby(['County', 'Radionuclide_1','Radionuclide_2','Radionuclide_3','Radionuclide_4','Radionuclide_5','Radionuclide_6']).size().reset_index(name='Count')
```

```python
df_isotopes = pd.DataFrame({'Isotope': isotopes})
```

```python
df_isotopes
```

```python
for isotope in df_isotopes['Isotope']:
    for col in county_isotope_count.columns[1:]:
        if county_isotope_count[col].dtype == 'object':
            county_list = county_isotope_count[county_isotope_count[col].str.contains(isotope, na=False)]['County'].tolist()
            if county_list:
                for county in county_list:
                    print(f'{isotope} is present in {county} in {col}')
        else:
            continue
```

```python
results = []
for isotope in df_isotopes['Isotope']:
    for col in county_isotope_count.columns[1:]:
        county_list = county_isotope_count[county_isotope_count[col].astype(str).str.contains(isotope)]['County'].tolist()
        if county_list:
            for county in county_list:
                results.append((isotope, county))


df_results = pd.DataFrame(results, columns=['Isotope', 'County'])
print(df_results)
```
